## Problema 1

```python
def problema1(imagen, color):
    #La imagen debe venir previamente leída por cv2
    img = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
    imgn = img.copy()
    if int(color) == 1:
        # cambiamos los canales rojo y verde a 0
        imgn[:, :, 0] = 0
        imgn[:, :, 1] = 0
    
    elif int(color) == 2:
        # cambiamos los canales rojo y azul a 0
        imgn[:, :, 0] = 0
        imgn[:, :, 2] = 0
    
    elif int(color) == 3:
        #cambiamos los canales verde y azul a 0
        imgn[:, :, 1] = 0
        imgn[:, :, 2] = 0
        
    elif int(color) == 10:
        #cambiamos el canal azul a 0
        imgn[:, :, 2] = 0
        
    elif int(color) == 20:
         #cambiamos el canal rojo a 0
        imgn[:, :, 0] = 0
        
    elif int(color) == 30:
         #cambiamos el canal verde a 0
        imgn[:, :, 1] = 0
    else:
        raise Exception('color no válido')
        
    plt.imshow(imgn)
    plt.show()
```

```python
def problema2(rojo,verde,azul):
    #las imágenes tienen que venir leídas por cv2
    img_r = rojo.copy()
    img_v = verde.copy()
    img_b = azul.copy()
    
    #cambiamos los canales verde y azul a 0 en la imagen rojo
    img_r[:, :, 1] = 0
    img_r[:, :, 2] = 0
    
    #cambiamos los canales rojo y azul a 0 en la imagen verde
    img_v[:, :, 0] = 0
    img_v[:, :, 2] = 0
    
    #cambiamos los canales rojo y verde a 0 en la imagen azul
    img_b[:, :, 0] = 0
    img_b[:, :, 1] = 0
    
    #Sumamos las tres imágenes resultantes
    imagen = img_r + img_v + img_b
    
    return imagen
   # plt.imshow(imagen)
  #  plt.show()
```

```python
def problema3(imagen):
    #La imagen debe venir leída por cv2
    img_r = imagen.copy()
    img_v = imagen.copy()
    img_b = imagen.copy()
    
    #Cambiamos los canales verde y azul al mismo valor del canal rojo
    img_r[:, :, 1] = img_r[:, :, 0]
    img_r[:, :, 2] = img_r[:, :, 0]
    
    #Cambiamos los canales rojo y azul al mismo valor del canal verde
    img_v[:, :, 0] = img_v[:, :, 1]
    img_v[:, :, 2] = img_v[:, :, 1]
    
    #Cambiamos los canales verde y rojo al mismo valor del canal azul
    img_b[:, :, 0] = img_b[:, :, 2]
    img_b[:, :, 1] = img_b[:, :, 2]
    
    return img_r, img_v, img_b

    
```

## Problema 4

```python
def problema4(imagen):
    #la imagen debe venir leída por cv2
    image = imagen.copy()
    ancho = image.shape[0]
    alto = image.shape[1]
    gray = np.zeros((ancho, alto, 4))

    for i in range(0, ancho):
        for j in range(0, alto):
            pixel = image[i, j]
            aritmetico = int((int(pixel[0]) + int(pixel[1]) + int(pixel[2]))/3)
            gray[i,j] = np.append(pixel,np.array(aritmetico))
    gray = gray.astype(np.uint8)
    color = ('b','g','r','gray')
    plt.figure(figsize=(16, 4))
    for i,col in enumerate(color):
        piv_plot = int(str(14)+str(i+1))
        histr = cv2.calcHist([gray],[i],None,[256],[0,256])
        plt.subplot(piv_plot)
        plt.plot(histr,color = col)
        plt.hlines(np.mean(histr),color = col,xmin=0, xmax=256)
        plt.xlim([0,256])
    plt.show()
```

## Problema 5

Las imágenes en escala de grises pueden ser el resultado de medir la intensidad de la luz en cada píxel según una combinación ponderada particular de frecuencias o longitudes de onda, y en tales casos son monocromáticas cuando solo una frecuencia está capturado.

La conversión de una imagen de color arbitraria a escala de grises no es única en general; la diferente ponderación de los canales de color representa efectivamente el efecto de filmar película en blanco y negro con filtros fotográficos de diferentes colores en las cámaras.

Para convertir un color de un espacio de color basado en un modelo de color RGB típico gamma comprimido (no lineal) a una representación en escala de grises de su luminancia, primero se debe eliminar la función de compresión gamma mediante expansión gamma (linealización) para transformar la imagen en un RGB lineal espacio de color, para que la suma ponderada apropiada se pueda aplicar a los componentes de color lineales ${\displaystyle R_{\mathrm{lineal}}, G_{\mathrm {lineal}}, B_{\mathrm{lineal}}}$ ) para calcular la luminancia lineal ${\displaystyle Y_{\mathrm{lineal}}} $, que luego se puede volver a comprimir con gamma si el resultado de la escala de grises también debe codificarse y almacenarse en un espacio de color no lineal típico.

El espacio de color sRGB se define en términos de la luminancia lineal CIE 1931 Y lineal , que viene dada por

$${\displaystyle Y_{\mathrm{lineal}} = 0.2126R_{\mathrm{lineal}} + 0.7152G_{\mathrm{lineal}} + 0.0722B_{\mathrm{lineal}}}$$

Estos tres coeficientes particulares representan la percepción de la intensidad (luminancia) de humanos tricrómicos típicos a la luz de la Rec. 709 colores primarios aditivos (cromaticidades) que se usan en la definición de sRGB. La visión humana es más sensible al verde, por lo que tiene el mayor valor de coeficiente (0.7152) y es menos sensible al azul, por lo que tiene el menor coeficiente (0.0722). 

**Luma**

En una transmisión de señal de vídeo, luma es la componente que codifica la información de luminosidad de la imagen. En términos generales, es algo muy similar a la versión en blanco y negro de la imagen original. Luma y crominancia combinadas proporcionan la señal denominada señal de vídeo compuesto, utilizada en multitud de aplicaciones; pueden enviarse conjuntamente o transmitirse independientemente. Forman parte de la codificación de vídeo en los estándares de TV NTSC y PAL, entre otros.

En los formatos digitales que siguen el estándar CCIR_601, la luma ${\displaystyle Y'}$ de un píxel se calcula con la fórmula 

$${\displaystyle Y'=0,299R'+0,587G'+0,114B'}$$

Fuente: https://www.hisour.com/es/grayscale-26279/

```python
def problema5(imagen):
     #la imagen debe venir leída por cv2
    image = imagen.copy()
    ancho = image.shape[0]
    alto = image.shape[1]

    gray = np.zeros((ancho, alto, 1))

    for i in range(0, ancho):
        for j in range(0, alto):
            pixel = image[i, j]
            ponderado = (int(0.299*pixel[0]) + int(0.587*pixel[1]) + int(0.114*pixel[2]))
            gray[i,j] = ponderado
    gray = gray.astype(np.uint8)
    return gray
```

```python
grises = problema5(imagen3)
```

```python
plt.imshow(grises, cmap='gray')
plt.show()
```

## Problema 6

### Espacio de color HSV

Sus siglas indican Hue Saturation Value y al igual que RGB es un mapeo de los distintos colores que son captados por los huanos pero en un arreglo cilindico en lugar de cúbico con las coordenadas como siguen:

- $\varphi$ : Es indicado por el Hue, es decir el color en sí, donde el color rojo se denota con $0^{\circ}$, el verde con $120^{\circ}$ y el azul con $240^{\circ}$, siendo los grados intermedios entre cada dos de éstos tres el color formado con la proporción de esos dos que tiene el color
- $\rho$ : Es indicado por la Saturation, es decir el colorido de un estímulo en relación con su propio brillo o la cantidad de luz blanca en un Hue (color). 
- $Z$ : Es el indicado por el Value, es decir la noción  cromática de intensidad, mientras sea pequeño el color es similar al blanco y mientras sea grande el color es similar al Hue.

Lo cual forma el sigueinte cilindro: <img src='.\HSV_color_solid_cylinder_saturation_gray.png'>
Fuente: [https://en.wikipedia.org/wiki/HSL_and_HSV](https://en.wikipedia.org/wiki/HSL_and_HSV).

Éste espacio fe diseñado en los años 1970s por investigadores que querían un espacio más allineado con la forma en que los atributos que crean colores son percibidos por los humanos.
